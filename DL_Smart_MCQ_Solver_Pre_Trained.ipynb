```python

# STEP 1
# Setup + Configuration + Data Loading
# Google Colab | Tesla T4 | PyTorch 2.11


import os
import gc
import math
import random
import warnings

import numpy as np
import pandas as pd

import torch
import torch.nn as nn

from tqdm.auto import tqdm

from sklearn.model_selection import train_test_split

from transformers import (
    AutoTokenizer,
    AutoModelForMultipleChoice,
    get_cosine_schedule_with_warmup
)

from torch.utils.data import Dataset, DataLoader

warnings.filterwarnings("ignore")


# Reproducibility


SEED = 42

random.seed(SEED)
np.random.seed(SEED)

torch.manual_seed(SEED)

if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)


# Device


DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("=" * 60)
print("Torch          :", torch.__version__)
print("CUDA           :", torch.version.cuda)

import transformers

print("Transformers   :", transformers.__version__)

if torch.cuda.is_available():

    print("GPU            :", torch.cuda.get_device_name(0))

    print(
        "GPU Memory     : %.2f GB"
        % (
            torch.cuda.get_device_properties(0).total_memory
            / 1024**3
        )
    )

print("=" * 60)


# Performance


torch.backends.cuda.matmul.allow_tf32 = True
torch.backends.cudnn.allow_tf32 = True
torch.backends.cudnn.benchmark = True

torch.set_float32_matmul_precision("high")


# Configuration


MODEL_NAME = "microsoft/deberta-v3-base"

MAX_LENGTH = 320

TRAIN_BATCH_SIZE = 1

VALID_BATCH_SIZE = 2

GRAD_ACCUM = 16

EPOCHS = 4

LR = 2e-5

WEIGHT_DECAY = 0.01

WARMUP_RATIO = 0.10

BEST_MODEL_PATH = "best_model.pt"

print("\nModel :", MODEL_NAME)


# Tokenizer


tokenizer = AutoTokenizer.from_pretrained(
    MODEL_NAME
)

print("Tokenizer Loaded")


# File Paths


TRAIN_PATH = "/content/train.csv"
TEST_PATH = "/content/test.csv"
SAMPLE_PATH = "/content/sample_submission.csv"


# Load Data


train_df = pd.read_csv(TRAIN_PATH)
test_df = pd.read_csv(TEST_PATH)
sample_submission_df = pd.read_csv(SAMPLE_PATH)

print()

print("Train Shape :", train_df.shape)
print("Test Shape  :", test_df.shape)

display(train_df.head())


# Fill Missing


TEXT_COLUMNS = [
    "prompt",
    "A",
    "B",
    "C",
    "D",
    "E"
]

for col in TEXT_COLUMNS:

    train_df[col] = (
        train_df[col]
        .fillna("")
        .astype(str)
        .str.strip()
    )

    test_df[col] = (
        test_df[col]
        .fillna("")
        .astype(str)
        .str.strip()
    )


# Labels


LABEL2ID = {
    "A": 0,
    "B": 1,
    "C": 2,
    "D": 3,
    "E": 4
}

ID2LABEL = {v: k for k, v in LABEL2ID.items()}

train_df["label"] = train_df["answer"].map(LABEL2ID)

print()

print(train_df["label"].value_counts().sort_index())

# Shuffle

train_df = train_df.sample(
    frac=1,
    random_state=SEED
).reset_index(drop=True)

# Train / Validation Split

train_data, valid_data = train_test_split(

    train_df,

    test_size=0.10,

    random_state=SEED,

    stratify=train_df["label"]

)

train_data = train_data.reset_index(drop=True)
valid_data = valid_data.reset_index(drop=True)

print()

print("Training Samples  :", len(train_data))
print("Validation Samples:", len(valid_data))
print("Test Samples      :", len(test_df))

# Cleanup

gc.collect()

if torch.cuda.is_available():
    torch.cuda.empty_cache()

print("\n STEP 1 Complete.")
```

```python

# STEP 2
# Fast Dataset (Pre-tokenized)


from torch.utils.data import Dataset, DataLoader
import os

# Dataset

class MCQDataset(Dataset):

    def __init__(self, dataframe, tokenizer, max_length=320, train=True):

        self.train = train
        self.samples = []

        print("Tokenizing dataset...")

        for _, row in tqdm(
            dataframe.iterrows(),
            total=len(dataframe)
        ):

            prompt = str(row["prompt"])

            choices = [
                str(row["A"]),
                str(row["B"]),
                str(row["C"]),
                str(row["D"]),
                str(row["E"])
            ]

            encoding = tokenizer(

                [prompt] * 5,

                choices,

                truncation=True,

                max_length=max_length,

                padding=False,

                return_attention_mask=True

            )

            sample = {

                "input_ids": encoding["input_ids"],

                "attention_mask": encoding["attention_mask"]

            }

            if "token_type_ids" in encoding:

                sample["token_type_ids"] = encoding["token_type_ids"]

            if train:

                sample["labels"] = int(row["label"])

            self.samples.append(sample)

        print("Finished Tokenizing.")

    def __len__(self):

        return len(self.samples)

    def __getitem__(self, idx):

        return self.samples[idx]


# Dynamic Padding

def collate_fn(features):

    labels = None

    if "labels" in features[0]:

        labels = torch.tensor(
            [x["labels"] for x in features],
            dtype=torch.long
        )

    batch_size = len(features)

    num_choices = len(features[0]["input_ids"])

    flattened = []

    for feature in features:

        for i in range(num_choices):

            item = {

                "input_ids":
                    feature["input_ids"][i],

                "attention_mask":
                    feature["attention_mask"][i]

            }

            if "token_type_ids" in feature:

                item["token_type_ids"] = (
                    feature["token_type_ids"][i]
                )

            flattened.append(item)

    batch = tokenizer.pad(

        flattened,

        padding="longest",

        pad_to_multiple_of=8,

        return_tensors="pt"

    )

    batch = {

        k: v.view(batch_size, num_choices, -1)

        for k, v in batch.items()

    }

    if labels is not None:

        batch["labels"] = labels

    return batch


# Build datasets

train_dataset = MCQDataset(

    train_data,

    tokenizer,

    MAX_LENGTH,

    train=True

)

valid_dataset = MCQDataset(

    valid_data,

    tokenizer,

    MAX_LENGTH,

    train=True

)

test_dataset = MCQDataset(

    test_df,

    tokenizer,

    MAX_LENGTH,

    train=False

)

print()

print("Train :", len(train_dataset))
print("Valid :", len(valid_dataset))
print("Test  :", len(test_dataset))

# DataLoader

NUM_WORKERS = min(4, os.cpu_count())

train_loader = DataLoader(

    train_dataset,

    batch_size=TRAIN_BATCH_SIZE,

    shuffle=True,

    collate_fn=collate_fn,

    pin_memory=True,

    num_workers=NUM_WORKERS,

    persistent_workers=(NUM_WORKERS > 0),

    prefetch_factor=2 if NUM_WORKERS > 0 else None,

    drop_last=True

)

valid_loader = DataLoader(

    valid_dataset,

    batch_size=VALID_BATCH_SIZE,

    shuffle=False,

    collate_fn=collate_fn,

    pin_memory=True,

    num_workers=NUM_WORKERS,

    persistent_workers=(NUM_WORKERS > 0),

    prefetch_factor=2 if NUM_WORKERS > 0 else None,

    drop_last=False

)

test_loader = DataLoader(

    test_dataset,

    batch_size=VALID_BATCH_SIZE,

    shuffle=False,

    collate_fn=collate_fn,

    pin_memory=True,

    num_workers=NUM_WORKERS,

    persistent_workers=(NUM_WORKERS > 0),

    prefetch_factor=2 if NUM_WORKERS > 0 else None,

    drop_last=False

)

print("\nDataLoaders Ready")

# Test batch

batch = next(iter(train_loader))

print("\nBatch Shapes")

for k, v in batch.items():

    print(f"{k:18} {tuple(v.shape)}")

gc.collect()

if torch.cuda.is_available():

    torch.cuda.empty_cache()

print("\nSTEP 2 Complete")
```

```python

# STEP 3
# Model + Optimizer + Scheduler + AMP + Validation (Debug)
# Compatible with Chunk 1 & Chunk 2


import gc
import numpy as np
import torch
import transformers

print("=" * 60)
print("Loading Model")
print("=" * 60)

print("Torch Version        :", torch.__version__)
print("Transformers Version :", transformers.__version__)
print("Default Dtype        :", torch.get_default_dtype())
print("Device               :", DEVICE)

# Load Model

model = AutoModelForMultipleChoice.from_pretrained(
    MODEL_NAME,
    ignore_mismatched_sizes=True
)

print("\nAfter from_pretrained()")
print(next(model.parameters()).dtype)

# Force FP32

model.float()

print("\nAfter model.float()")
print(next(model.parameters()).dtype)

# Gradient Checkpointing

model.gradient_checkpointing_enable()
model.config.use_cache = False

# Move to GPU

model.to(DEVICE)

print("\nAfter model.to()")
print(next(model.parameters()).dtype)

# Count parameter dtypes

fp16_params = 0
fp32_params = 0

for p in model.parameters():

    if p.dtype == torch.float16:
        fp16_params += 1

    elif p.dtype == torch.float32:
        fp32_params += 1

print("\nParameter Summary")
print("----------------------------")
print("FP16 Parameters :", fp16_params)
print("FP32 Parameters :", fp32_params)

# STOP IMMEDIATELY IF MODEL IS FP16

if fp16_params > 0:

    raise RuntimeError(
        "ERROR: Model parameters are FP16 before training.\n"
        "AMP requires FP32 model weights.\n"
        "Do NOT continue until this is fixed."
    )

print("\nModel verified as FP32.")

# Parameter Count

total_params = sum(
    p.numel()
    for p in model.parameters()
)

trainable_params = sum(
    p.numel()
    for p in model.parameters()
    if p.requires_grad
)

print("\nTotal Parameters :", f"{total_params:,}")
print("Trainable Params :", f"{trainable_params:,}")

# Optimizer

no_decay = ["bias", "LayerNorm.weight"]

optimizer_grouped_parameters = [

    {
        "params":[
            p for n,p in model.named_parameters()
            if not any(nd in n for nd in no_decay)
        ],
        "weight_decay":WEIGHT_DECAY
    },

    {
        "params":[
            p for n,p in model.named_parameters()
            if any(nd in n for nd in no_decay)
        ],
        "weight_decay":0.0
    }

]

optimizer = torch.optim.AdamW(

    optimizer_grouped_parameters,

    lr=LR,

    betas=(0.9,0.999),

    eps=1e-8

)

print("\nOptimizer Ready")

# Scheduler

steps_per_epoch = len(train_loader)

optimizer_steps_per_epoch = (
    steps_per_epoch + GRAD_ACCUM - 1
) // GRAD_ACCUM

total_training_steps = optimizer_steps_per_epoch * EPOCHS

warmup_steps = int(
    total_training_steps * WARMUP_RATIO
)

scheduler = get_cosine_schedule_with_warmup(

    optimizer,

    num_warmup_steps=warmup_steps,

    num_training_steps=total_training_steps

)

print("Scheduler Ready")
print("Total Steps :", total_training_steps)
print("Warmup Steps:", warmup_steps)

# AMP

scaler = torch.amp.GradScaler("cuda")

print("\nGradScaler Enabled :", scaler.is_enabled())

# mAP@3

def map3(logits, labels):

    top3 = np.argsort(-logits, axis=1)[:, :3]

    score = 0.0

    for truth, pred in zip(labels, top3):

        if truth in pred:

            rank = np.where(pred == truth)[0][0] + 1

            score += 1.0 / rank

    return score / len(labels)

# Validation

@torch.no_grad()
def validate():

    model.eval()

    losses = []
    preds = []
    labels = []

    for batch in valid_loader:

        batch = {
            k: v.to(DEVICE, non_blocking=True)
            for k, v in batch.items()
        }

        with torch.amp.autocast("cuda"):

            outputs = model(**batch)

        losses.append(outputs.loss.item())

        preds.append(
            outputs.logits.float().cpu().numpy()
        )

        labels.append(
            batch["labels"].cpu().numpy()
        )

    preds = np.concatenate(preds)
    labels = np.concatenate(labels)

    acc = (
        preds.argmax(1) == labels
    ).mean()

    score = map3(preds, labels)

    return (
        float(np.mean(losses)),
        float(acc),
        float(score)
    )

gc.collect()

if torch.cuda.is_available():
    torch.cuda.empty_cache()

print("\nSTEP 3 Complete.")
```

```python

# STEP 4
# Training Loop


import time

best_map3 = -1.0

history = {
    "train_loss": [],
    "valid_loss": [],
    "valid_acc": [],
    "valid_map3": []
}

print("=" * 70)
print("Starting Training")
print("=" * 70)

for epoch in range(EPOCHS):

    print()
    print("=" * 70)
    print(f"Epoch {epoch + 1}/{EPOCHS}")
    print("=" * 70)

    model.train()

    running_loss = 0.0

    optimizer.zero_grad(set_to_none=True)

    progress = tqdm(
        enumerate(train_loader),
        total=len(train_loader),
        leave=True
    )

    start_time = time.time()

    for step, batch in progress:

        batch = {
            k: v.to(DEVICE, non_blocking=True)
            for k, v in batch.items()
        }

        with torch.amp.autocast("cuda"):

            outputs = model(**batch)

            loss = outputs.loss

            loss = loss / GRAD_ACCUM

        scaler.scale(loss).backward()

        running_loss += loss.item() * GRAD_ACCUM

        if (
            (step + 1) % GRAD_ACCUM == 0
            or
            (step + 1) == len(train_loader)
        ):

            scaler.unscale_(optimizer)

            torch.nn.utils.clip_grad_norm_(
                model.parameters(),
                1.0
            )

            scaler.step(optimizer)

            scaler.update()

            scheduler.step()

            optimizer.zero_grad(set_to_none=True)

        progress.set_postfix(
            loss=f"{running_loss/(step+1):.4f}",
            lr=f"{scheduler.get_last_lr()[0]:.2e}"
        )

    epoch_time = time.time() - start_time

    train_loss = running_loss / len(train_loader)

    valid_loss, valid_acc, valid_map3 = validate()

    history["train_loss"].append(train_loss)
    history["valid_loss"].append(valid_loss)
    history["valid_acc"].append(valid_acc)
    history["valid_map3"].append(valid_map3)

    print()

    print(f"Epoch Time      : {epoch_time:.1f} sec")
    print(f"Train Loss      : {train_loss:.4f}")
    print(f"Validation Loss : {valid_loss:.4f}")
    print(f"Validation Acc  : {valid_acc:.4f}")
    print(f"Validation mAP3 : {valid_map3:.4f}")

    if valid_map3 > best_map3:

        best_map3 = valid_map3

        torch.save(
            model.state_dict(),
            BEST_MODEL_PATH
        )

        print("Best model saved.")

print()
print("=" * 70)
print("Training Finished")
print("=" * 70)

print(f"Best Validation mAP@3 : {best_map3:.5f}")
```

```python

# STEP 5
# Inference + submission.csv


import numpy as np
import pandas as pd
import torch
from tqdm.auto import tqdm

# Load best model

model.load_state_dict(
    torch.load(BEST_MODEL_PATH, map_location=DEVICE)
)

model.to(DEVICE)
model.eval()

print("Best model loaded.")

# Prediction

all_predictions = []

with torch.no_grad():

    for batch in tqdm(test_loader):

        batch = {
            k: v.to(DEVICE)
            for k, v in batch.items()
        }

        with torch.amp.autocast("cuda"):

            logits = model(**batch).logits

        all_predictions.append(
            logits.float().cpu().numpy()
        )

all_predictions = np.concatenate(all_predictions)

print(all_predictions.shape)

# Top-3

top3 = np.argsort(-all_predictions, axis=1)[:, :3]

predictions = []

for row in top3:

    predictions.append(
        " ".join(
            ID2LABEL[i]
            for i in row
        )
    )

# Submission

submission = sample_submission_df.copy()

submission["Prediction"] = predictions

submission.to_csv(
    "submission.csv",
    index=False
)

print("\nsubmission.csv created.")

display(submission.head())
```

```python
# Step 1. Install libraries (Kaggle/Colab)
!pip install -q transformers datasets accelerate evaluate sentencepiece

# Step 2 — Import Libraries
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import torch

from datasets import Dataset

from transformers import (
    AutoTokenizer,
    AutoModelForMultipleChoice,
    TrainingArguments,
    Trainer
)

from sklearn.model_selection import train_test_split

# Step 3 — Load Dataset

train_df = pd.read_csv("/content/train.csv")
test_df = pd.read_csv("/content/test.csv")
sample_submission = pd.read_csv("/content/sample_submission.csv")

print(train_df.shape)
print(test_df.shape)

# Step 4 — Look at the Dataset

train_df.head()

# Step 5 — Convert Answers to Numbers

label2id = {
    "A":0,
    "B":1,
    "C":2,
    "D":3,
    "E":4
}

id2label = {v:k for k,v in label2id.items()}

train_df["label"] = train_df["answer"].map(label2id)

train_df[["answer","label"]].head()

# Step 6 — Train Validation Split
train_df, valid_df = train_test_split(
    train_df,
    test_size=0.1,
    random_state=42,
    stratify=train_df["label"]
)
print(train_df.shape)
print(valid_df.shape)

# Step 7 — Convert to HuggingFace Dataset
train_dataset = Dataset.from_pandas(train_df)
valid_dataset = Dataset.from_pandas(valid_df)
test_dataset = Dataset.from_pandas(test_df)

# Step 8 — Download ELECTRA Base
MODEL_NAME = "google/electra-base-discriminator"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

# Step 9 — Multiple Choice Preprocessing
def preprocess_function(example):

    first_sentences = [example["prompt"]] * 5

    second_sentences = [
        example["A"],
        example["B"],
        example["C"],
        example["D"],
        example["E"]
    ]

    tokenized = tokenizer(
        first_sentences,
        second_sentences,
        truncation=True,
        max_length=384
    )

    tokenized["label"] = example["label"]

    return tokenized

print(train_dataset.column_names)

from datasets import Dataset

train_dataset = Dataset.from_pandas(train_df, preserve_index=False)
valid_dataset = Dataset.from_pandas(valid_df, preserve_index=False)
test_dataset = Dataset.from_pandas(test_df, preserve_index=False)

print(train_dataset.column_names)

```

```python
# Step 10 — Preprocess Training Data
train_dataset = train_dataset.map(preprocess_function)

# Step 11 — Validation Data
valid_dataset = valid_dataset.map(preprocess_function)

# Step 12 — Test Data
def preprocess_test(example):

    first_sentences = [example["prompt"]] * 5

    second_sentences = [
        example["A"],
        example["B"],
        example["C"],
        example["D"],
        example["E"]
    ]

    return tokenizer(
        first_sentences,
        second_sentences,
        truncation=True,
        max_length=384
    )

test_dataset = test_dataset.map(preprocess_test)

# Step 13 — Verify Everything Worked
print(train_dataset[0].keys())
```

```python
# Step 14 Remove unwanted columns safely

train_remove = [
    col for col in [
        "id", "prompt", "A", "B", "C", "D", "E",
        "answer", "__index_level_0__"
    ]
    if col in train_dataset.column_names
]

valid_remove = [
    col for col in [
        "id", "prompt", "A", "B", "C", "D", "E",
        "answer", "__index_level_0__"
    ]
    if col in valid_dataset.column_names
]

test_remove = [
    col for col in [
        "id", "prompt", "A", "B", "C", "D", "E",
        "__index_level_0__"
    ]
    if col in test_dataset.column_names
]

train_dataset = train_dataset.remove_columns(train_remove)
valid_dataset = valid_dataset.remove_columns(valid_remove)
test_dataset = test_dataset.remove_columns(test_remove)

print(train_dataset.column_names)
print(valid_dataset.column_names)
print(test_dataset.column_names)


# Step 15 - Check the processed dataset
print(train_dataset[0])

# Step 16 - Load ELECTRA Base
MODEL_NAME = "google/electra-base-discriminator"

model = AutoModelForMultipleChoice.from_pretrained(MODEL_NAME)

# Step 17 - Create Data Collator
from dataclasses import dataclass
from typing import Optional, Union
import torch

from transformers import PreTrainedTokenizerBase
from transformers.tokenization_utils_base import PaddingStrategy

@dataclass
class DataCollatorForMultipleChoice:
    tokenizer: PreTrainedTokenizerBase
    padding: Union[bool, str, PaddingStrategy] = True
    max_length: Optional[int] = None

    def __call__(self, features):

        labels = None
        if "label" in features[0]:
            labels = [feature.pop("label") for feature in features]

        batch_size = len(features)
        num_choices = len(features[0]["input_ids"])

        flattened_features = []

        for feature in features:
            for i in range(num_choices):
                flattened_features.append(
                    {k: v[i] for k, v in feature.items()}
                )

        batch = self.tokenizer.pad(
            flattened_features,
            padding=self.padding,
            max_length=self.max_length,
            return_tensors="pt"
        )

        batch = {
            k: v.view(batch_size, num_choices, -1)
            for k, v in batch.items()
        }

        if labels is not None:
            batch["labels"] = torch.tensor(labels)

        return batch


# Step 18 - Initialize Data Collator
data_collator = DataCollatorForMultipleChoice(tokenizer)

# Step 19 - Training Arguments
training_args = TrainingArguments(
    output_dir="./electra_model",

    learning_rate=1e-5,

    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,

    gradient_accumulation_steps=2,

    num_train_epochs=5,

    warmup_ratio=0.1,

    weight_decay=0.05,

    logging_steps=50,

    eval_strategy="epoch",

    save_strategy="epoch",

    load_best_model_at_end=True,

    metric_for_best_model="eval_loss",

    greater_is_better=False,

    report_to="none",

    fp16=torch.cuda.is_available()
)


# Step 20 - Create Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=valid_dataset,
    data_collator=data_collator
)
```

```python
# Step 21 - Train
trainer.train()
```

```python
# Step 22 - Evaluate on the Validation Set
eval_results = trainer.evaluate()

print(eval_results)

# Step 23 - Predict on the Test Set
test_predictions = trainer.predict(test_dataset)

# Step 24 - Check prediction shape
logits = test_predictions.predictions

print(logits.shape)

# Step 25 - Get the predicted answer (Top-1)
pred_ids = np.argmax(logits, axis=1)

id2label = {
    0: "A",
    1: "B",
    2: "C",
    3: "D",
    4: "E"
}

pred_answers = [id2label[i] for i in pred_ids]


# Step 26 - Create submission
submission = pd.DataFrame({
    "ID": test_df["id"],
    "Prediction": pred_answers
})

submission.head()


# Step 27 - Save submission
submission.to_csv("ELECTRA submission.csv", index=False)

print("ELECTRA submission.csv created successfully!")
```
