```python
# Section 1: Install & Import Libraries
!pip install gensim

import re
import string
import numpy as np
import pandas as pd

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

from gensim.models import Word2Vec

# Download NLTK resources
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
```

```python
# Section 2: Text Cleaning, Tokenization & Missing Data Handling

def clean_text(text):

    # Handle missing values
    if not isinstance(text, str):
        text = ''

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = re.sub(
        f'[{re.escape(string.punctuation)}]',
        '',
        text
    )

    # Tokenization
    tokens = word_tokenize(text)

    # Stopword removal
    stop_words = set(stopwords.words('english'))

    filtered_tokens = [
        word
        for word in tokens
        if word not in stop_words and word.isalpha()
    ]

    return ' '.join(filtered_tokens), filtered_tokens
```

```python
import nltk

# Download required NLTK resources
resources = [
    "punkt",
    "punkt_tab",
    "stopwords",
    "wordnet",
    "omw-1.4"
]

for resource in resources:
    nltk.download(resource)

# Section 3: Apply Cleaning to Train and Test Data
text_cols = ['prompt', 'A', 'B', 'C', 'D', 'E']

print("Cleaning text...")

for df in [train_df, test_df]:
    for col in text_cols:
        df[col] = df[col].fillna('')

        cleaned = df[col].apply(clean_text)

        df[f'{col}_cleaned'] = cleaned.apply(lambda x: x[0])
        df[f'{col}_tokens'] = cleaned.apply(lambda x: x[1])

print("Cleaning completed.")
```

```python
# Section 4: Generate TF-IDF Embeddings

print("Generating TF-IDF embeddings...")

all_cleaned_text = []

for col in [f'{c}_cleaned' for c in text_cols]:

    all_cleaned_text.extend(
        train_df[col].tolist()
    )

# Fit only on training data
tfidf_vectorizer = TfidfVectorizer(
    max_features=5000
)

tfidf_vectorizer.fit(all_cleaned_text)

# Train data embeddings
train_df['prompt_tfidf'] = list(
    tfidf_vectorizer.transform(
        train_df['prompt_cleaned']
    )
)

for col in ['A', 'B', 'C', 'D', 'E']:

    train_df[f'{col}_tfidf'] = list(
        tfidf_vectorizer.transform(
            train_df[f'{col}_cleaned']
        )
    )

# Test data embeddings
test_df['prompt_tfidf'] = list(
    tfidf_vectorizer.transform(
        test_df['prompt_cleaned']
    )
)

for col in ['A', 'B', 'C', 'D', 'E']:

    test_df[f'{col}_tfidf'] = list(
        tfidf_vectorizer.transform(
            test_df[f'{col}_cleaned']
        )
    )

print("TF-IDF completed.")
```

```python
# Section 5: Generate Word2Vec Embeddings

print("Training Word2Vec model...")

all_tokens = []

for col in [f'{c}_tokens' for c in text_cols]:

    all_tokens.extend(
        train_df[col].tolist()
    )

word2vec_model = Word2Vec(
    sentences=all_tokens,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)

print("Word2Vec training completed.")
```

```python
# Section 6: Convert Sentences to Word2Vec Embeddings
def get_word2vec_embedding(
    tokens,
    model,
    vector_size=100
):

    vectors = []

    for token in tokens:

        if token in model.wv:
            vectors.append(
                model.wv[token]
            )

    if len(vectors) > 0:

        return np.mean(
            vectors,
            axis=0
        )

    return np.zeros(vector_size)
```

```python
# Section 7: Apply Word2Vec Embeddings
train_df['prompt_w2v'] = train_df[
    'prompt_tokens'
].apply(
    lambda x:
    get_word2vec_embedding(
        x,
        word2vec_model
    )
)

for col in ['A', 'B', 'C', 'D', 'E']:

    train_df[f'{col}_w2v'] = train_df[
        f'{col}_tokens'
    ].apply(
        lambda x:
        get_word2vec_embedding(
            x,
            word2vec_model
        )
    )

test_df['prompt_w2v'] = test_df[
    'prompt_tokens'
].apply(
    lambda x:
    get_word2vec_embedding(
        x,
        word2vec_model
    )
)

for col in ['A', 'B', 'C', 'D', 'E']:

    test_df[f'{col}_w2v'] = test_df[
        f'{col}_tokens'
    ].apply(
        lambda x:
        get_word2vec_embedding(
            x,
            word2vec_model
        )
    )

print("Word2Vec embeddings generated.")
```

```python
# Section 8: Cosine Similarity Function
def compute_cosine_similarity(
    vec1,
    vec2
):

    # TF-IDF sparse matrices
    if hasattr(vec1, "toarray"):

        return cosine_similarity(
            vec1,
            vec2
        )[0][0]

    # Word2Vec vectors
    return cosine_similarity(
        vec1.reshape(1, -1),
        vec2.reshape(1, -1)
    )[0][0]
```

```python
# Section 9: Example Similarity Calculation
print("\nExample Similarities")

tfidf_similarity = compute_cosine_similarity(
    train_df['prompt_tfidf'].iloc[0],
    train_df['A_tfidf'].iloc[0]
)

print(
    f"TF-IDF Similarity: {tfidf_similarity:.4f}"
)

w2v_similarity = compute_cosine_similarity(
    train_df['prompt_w2v'].iloc[0],
    train_df['A_w2v'].iloc[0]
)

print(
    f"Word2Vec Similarity: {w2v_similarity:.4f}"
)
```

```python
# Section 10: AP@3 Calculation
def calculate_ap_at_k(
    actual,
    predicted,
    k=3
):

    predicted = predicted[:k]

    for i, p in enumerate(predicted):

        if p == actual:

            return 1.0 / (i + 1)

    return 0.0
```

```python
# Section 11: mAP@3 Calculation
def calculate_map_k(
    df,
    embedding_type='tfidf',
    k=3
):

    options = [
        'A',
        'B',
        'C',
        'D',
        'E'
    ]

    ap_scores = []

    for _, row in df.iterrows():

        prompt_vec = row[
            f'prompt_{embedding_type}'
        ]

        actual_answer = row['answer']

        similarities = []

        for opt in options:

            option_vec = row[
                f'{opt}_{embedding_type}'
            ]

            similarity = compute_cosine_similarity(
                prompt_vec,
                option_vec
            )

            similarities.append(
                (
                    opt,
                    similarity
                )
            )

        similarities.sort(
            key=lambda x: x[1],
            reverse=True
        )

        predicted_labels = [
            opt
            for opt, sim
            in similarities
        ]

        ap = calculate_ap_at_k(
            actual_answer,
            predicted_labels,
            k
        )

        ap_scores.append(ap)

    return np.mean(ap_scores)
```

```python
# Section 12: Calculate Final mAP@3 Scores
print("\nCalculating mAP@3")

map3_tfidf = calculate_map_k(
    train_df,
    embedding_type='tfidf',
    k=3
)

print(
    f"TF-IDF mAP@3: {map3_tfidf:.4f}"
)

map3_w2v = calculate_map_k(
    train_df,
    embedding_type='w2v',
    k=3
)

print(
    f"Word2Vec mAP@3: {map3_w2v:.4f}"
)
```

```python
# ============================================================
# CHUNK 1
# Setup + Configuration + Data Loading
# Google Colab | Tesla T4 | PyTorch 2.11
# ============================================================

import os
import gc
import math
import random
import warnings

import numpy as np
import pandas as pd

import torch
import torch.nn as nn

from tqdm.auto import tqdm

from sklearn.model_selection import train_test_split

from transformers import (
    AutoTokenizer,
    AutoModelForMultipleChoice,
    get_cosine_schedule_with_warmup
)

from torch.utils.data import Dataset, DataLoader

warnings.filterwarnings("ignore")

# ============================================================
# Reproducibility
# ============================================================

SEED = 42

random.seed(SEED)
np.random.seed(SEED)

torch.manual_seed(SEED)

if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ============================================================
# Device
# ============================================================

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("=" * 60)
print("Torch          :", torch.__version__)
print("CUDA           :", torch.version.cuda)

import transformers

print("Transformers   :", transformers.__version__)

if torch.cuda.is_available():

    print("GPU            :", torch.cuda.get_device_name(0))

    print(
        "GPU Memory     : %.2f GB"
        % (
            torch.cuda.get_device_properties(0).total_memory
            / 1024**3
        )
    )

print("=" * 60)

# ============================================================
# Performance
# ============================================================

torch.backends.cuda.matmul.allow_tf32 = True
torch.backends.cudnn.allow_tf32 = True
torch.backends.cudnn.benchmark = True

torch.set_float32_matmul_precision("high")

# ============================================================
# Configuration
# ============================================================

MODEL_NAME = "microsoft/deberta-v3-base"

MAX_LENGTH = 320

TRAIN_BATCH_SIZE = 1

VALID_BATCH_SIZE = 2

GRAD_ACCUM = 16

EPOCHS = 4

LR = 2e-5

WEIGHT_DECAY = 0.01

WARMUP_RATIO = 0.10

BEST_MODEL_PATH = "best_model.pt"

print("\nModel :", MODEL_NAME)

# ============================================================
# Tokenizer
# ============================================================

tokenizer = AutoTokenizer.from_pretrained(
    MODEL_NAME
)

print("Tokenizer Loaded")

# ============================================================
# File Paths
# ============================================================

TRAIN_PATH = "/content/train.csv"
TEST_PATH = "/content/test.csv"
SAMPLE_PATH = "/content/sample_submission.csv"

# ============================================================
# Load Data
# ============================================================

train_df = pd.read_csv(TRAIN_PATH)
test_df = pd.read_csv(TEST_PATH)
sample_submission_df = pd.read_csv(SAMPLE_PATH)

print()

print("Train Shape :", train_df.shape)
print("Test Shape  :", test_df.shape)

display(train_df.head())

# ============================================================
# Fill Missing
# ============================================================

TEXT_COLUMNS = [
    "prompt",
    "A",
    "B",
    "C",
    "D",
    "E"
]

for col in TEXT_COLUMNS:

    train_df[col] = (
        train_df[col]
        .fillna("")
        .astype(str)
        .str.strip()
    )

    test_df[col] = (
        test_df[col]
        .fillna("")
        .astype(str)
        .str.strip()
    )

# ============================================================
# Labels
# ============================================================

LABEL2ID = {
    "A": 0,
    "B": 1,
    "C": 2,
    "D": 3,
    "E": 4
}

ID2LABEL = {v: k for k, v in LABEL2ID.items()}

train_df["label"] = train_df["answer"].map(LABEL2ID)

print()

print(train_df["label"].value_counts().sort_index())

# ============================================================
# Shuffle
# ============================================================

train_df = train_df.sample(
    frac=1,
    random_state=SEED
).reset_index(drop=True)

# ============================================================
# Train / Validation Split
# ============================================================

train_data, valid_data = train_test_split(

    train_df,

    test_size=0.10,

    random_state=SEED,

    stratify=train_df["label"]

)

train_data = train_data.reset_index(drop=True)
valid_data = valid_data.reset_index(drop=True)

print()

print("Training Samples  :", len(train_data))
print("Validation Samples:", len(valid_data))
print("Test Samples      :", len(test_df))

# ============================================================
# Cleanup
# ============================================================

gc.collect()

if torch.cuda.is_available():
    torch.cuda.empty_cache()

print("\nChunk 1 Complete.")
```

```python
# ============================================================
# CHUNK 2
# Fast Dataset (Pre-tokenized)
# ============================================================

from torch.utils.data import Dataset, DataLoader
import os

# ------------------------------------------------------------
# Dataset
# ------------------------------------------------------------

class MCQDataset(Dataset):

    def __init__(self, dataframe, tokenizer, max_length=320, train=True):

        self.train = train
        self.samples = []

        print("Tokenizing dataset...")

        for _, row in tqdm(
            dataframe.iterrows(),
            total=len(dataframe)
        ):

            prompt = str(row["prompt"])

            choices = [
                str(row["A"]),
                str(row["B"]),
                str(row["C"]),
                str(row["D"]),
                str(row["E"])
            ]

            encoding = tokenizer(

                [prompt] * 5,

                choices,

                truncation=True,

                max_length=max_length,

                padding=False,

                return_attention_mask=True

            )

            sample = {

                "input_ids": encoding["input_ids"],

                "attention_mask": encoding["attention_mask"]

            }

            if "token_type_ids" in encoding:

                sample["token_type_ids"] = encoding["token_type_ids"]

            if train:

                sample["labels"] = int(row["label"])

            self.samples.append(sample)

        print("Finished Tokenizing.")

    def __len__(self):

        return len(self.samples)

    def __getitem__(self, idx):

        return self.samples[idx]


# ------------------------------------------------------------
# Dynamic Padding
# ------------------------------------------------------------

def collate_fn(features):

    labels = None

    if "labels" in features[0]:

        labels = torch.tensor(
            [x["labels"] for x in features],
            dtype=torch.long
        )

    batch_size = len(features)

    num_choices = len(features[0]["input_ids"])

    flattened = []

    for feature in features:

        for i in range(num_choices):

            item = {

                "input_ids":
                    feature["input_ids"][i],

                "attention_mask":
                    feature["attention_mask"][i]

            }

            if "token_type_ids" in feature:

                item["token_type_ids"] = (
                    feature["token_type_ids"][i]
                )

            flattened.append(item)

    batch = tokenizer.pad(

        flattened,

        padding="longest",

        pad_to_multiple_of=8,

        return_tensors="pt"

    )

    batch = {

        k: v.view(batch_size, num_choices, -1)

        for k, v in batch.items()

    }

    if labels is not None:

        batch["labels"] = labels

    return batch


# ------------------------------------------------------------
# Build datasets
# ------------------------------------------------------------

train_dataset = MCQDataset(

    train_data,

    tokenizer,

    MAX_LENGTH,

    train=True

)

valid_dataset = MCQDataset(

    valid_data,

    tokenizer,

    MAX_LENGTH,

    train=True

)

test_dataset = MCQDataset(

    test_df,

    tokenizer,

    MAX_LENGTH,

    train=False

)

print()

print("Train :", len(train_dataset))
print("Valid :", len(valid_dataset))
print("Test  :", len(test_dataset))

# ------------------------------------------------------------
# DataLoader
# ------------------------------------------------------------

NUM_WORKERS = min(4, os.cpu_count())

train_loader = DataLoader(

    train_dataset,

    batch_size=TRAIN_BATCH_SIZE,

    shuffle=True,

    collate_fn=collate_fn,

    pin_memory=True,

    num_workers=NUM_WORKERS,

    persistent_workers=(NUM_WORKERS > 0),

    prefetch_factor=2 if NUM_WORKERS > 0 else None,

    drop_last=True

)

valid_loader = DataLoader(

    valid_dataset,

    batch_size=VALID_BATCH_SIZE,

    shuffle=False,

    collate_fn=collate_fn,

    pin_memory=True,

    num_workers=NUM_WORKERS,

    persistent_workers=(NUM_WORKERS > 0),

    prefetch_factor=2 if NUM_WORKERS > 0 else None,

    drop_last=False

)

test_loader = DataLoader(

    test_dataset,

    batch_size=VALID_BATCH_SIZE,

    shuffle=False,

    collate_fn=collate_fn,

    pin_memory=True,

    num_workers=NUM_WORKERS,

    persistent_workers=(NUM_WORKERS > 0),

    prefetch_factor=2 if NUM_WORKERS > 0 else None,

    drop_last=False

)

print("\nDataLoaders Ready")

# ------------------------------------------------------------
# Test batch
# ------------------------------------------------------------

batch = next(iter(train_loader))

print("\nBatch Shapes")

for k, v in batch.items():

    print(f"{k:18} {tuple(v.shape)}")

gc.collect()

if torch.cuda.is_available():

    torch.cuda.empty_cache()

print("\nChunk 2 Complete")
```

```python
# ============================================================
# CHUNK 3
# Model + Optimizer + Scheduler + AMP + Validation (Debug)
# Compatible with Chunk 1 & Chunk 2
# ============================================================

import gc
import numpy as np
import torch
import transformers

print("=" * 60)
print("Loading Model")
print("=" * 60)

print("Torch Version        :", torch.__version__)
print("Transformers Version :", transformers.__version__)
print("Default Dtype        :", torch.get_default_dtype())
print("Device               :", DEVICE)

# ------------------------------------------------------------
# Load Model
# ------------------------------------------------------------

model = AutoModelForMultipleChoice.from_pretrained(
    MODEL_NAME,
    ignore_mismatched_sizes=True
)

print("\nAfter from_pretrained()")
print(next(model.parameters()).dtype)

# ------------------------------------------------------------
# Force FP32
# ------------------------------------------------------------

model.float()

print("\nAfter model.float()")
print(next(model.parameters()).dtype)

# ------------------------------------------------------------
# Gradient Checkpointing
# ------------------------------------------------------------

model.gradient_checkpointing_enable()
model.config.use_cache = False

# ------------------------------------------------------------
# Move to GPU
# ------------------------------------------------------------

model.to(DEVICE)

print("\nAfter model.to()")
print(next(model.parameters()).dtype)

# ------------------------------------------------------------
# Count parameter dtypes
# ------------------------------------------------------------

fp16_params = 0
fp32_params = 0

for p in model.parameters():

    if p.dtype == torch.float16:
        fp16_params += 1

    elif p.dtype == torch.float32:
        fp32_params += 1

print("\nParameter Summary")
print("----------------------------")
print("FP16 Parameters :", fp16_params)
print("FP32 Parameters :", fp32_params)

# ------------------------------------------------------------
# STOP IMMEDIATELY IF MODEL IS FP16
# ------------------------------------------------------------

if fp16_params > 0:

    raise RuntimeError(
        "ERROR: Model parameters are FP16 before training.\n"
        "AMP requires FP32 model weights.\n"
        "Do NOT continue until this is fixed."
    )

print("\nModel verified as FP32.")

# ------------------------------------------------------------
# Parameter Count
# ------------------------------------------------------------

total_params = sum(
    p.numel()
    for p in model.parameters()
)

trainable_params = sum(
    p.numel()
    for p in model.parameters()
    if p.requires_grad
)

print("\nTotal Parameters :", f"{total_params:,}")
print("Trainable Params :", f"{trainable_params:,}")

# ------------------------------------------------------------
# Optimizer
# ------------------------------------------------------------

no_decay = ["bias", "LayerNorm.weight"]

optimizer_grouped_parameters = [

    {
        "params":[
            p for n,p in model.named_parameters()
            if not any(nd in n for nd in no_decay)
        ],
        "weight_decay":WEIGHT_DECAY
    },

    {
        "params":[
            p for n,p in model.named_parameters()
            if any(nd in n for nd in no_decay)
        ],
        "weight_decay":0.0
    }

]

optimizer = torch.optim.AdamW(

    optimizer_grouped_parameters,

    lr=LR,

    betas=(0.9,0.999),

    eps=1e-8

)

print("\nOptimizer Ready")

# ------------------------------------------------------------
# Scheduler
# ------------------------------------------------------------

steps_per_epoch = len(train_loader)

optimizer_steps_per_epoch = (
    steps_per_epoch + GRAD_ACCUM - 1
) // GRAD_ACCUM

total_training_steps = optimizer_steps_per_epoch * EPOCHS

warmup_steps = int(
    total_training_steps * WARMUP_RATIO
)

scheduler = get_cosine_schedule_with_warmup(

    optimizer,

    num_warmup_steps=warmup_steps,

    num_training_steps=total_training_steps

)

print("Scheduler Ready")
print("Total Steps :", total_training_steps)
print("Warmup Steps:", warmup_steps)

# ------------------------------------------------------------
# AMP
# ------------------------------------------------------------

scaler = torch.amp.GradScaler("cuda")

print("\nGradScaler Enabled :", scaler.is_enabled())

# ------------------------------------------------------------
# mAP@3
# ------------------------------------------------------------

def map3(logits, labels):

    top3 = np.argsort(-logits, axis=1)[:, :3]

    score = 0.0

    for truth, pred in zip(labels, top3):

        if truth in pred:

            rank = np.where(pred == truth)[0][0] + 1

            score += 1.0 / rank

    return score / len(labels)

# ------------------------------------------------------------
# Validation
# ------------------------------------------------------------

@torch.no_grad()
def validate():

    model.eval()

    losses = []
    preds = []
    labels = []

    for batch in valid_loader:

        batch = {
            k: v.to(DEVICE, non_blocking=True)
            for k, v in batch.items()
        }

        with torch.amp.autocast("cuda"):

            outputs = model(**batch)

        losses.append(outputs.loss.item())

        preds.append(
            outputs.logits.float().cpu().numpy()
        )

        labels.append(
            batch["labels"].cpu().numpy()
        )

    preds = np.concatenate(preds)
    labels = np.concatenate(labels)

    acc = (
        preds.argmax(1) == labels
    ).mean()

    score = map3(preds, labels)

    return (
        float(np.mean(losses)),
        float(acc),
        float(score)
    )

gc.collect()

if torch.cuda.is_available():
    torch.cuda.empty_cache()

print("\nChunk 3 Complete.")
```

```python
# ============================================================
# CHUNK 4
# Training Loop
# ============================================================

import time

best_map3 = -1.0

history = {
    "train_loss": [],
    "valid_loss": [],
    "valid_acc": [],
    "valid_map3": []
}

print("=" * 70)
print("Starting Training")
print("=" * 70)

for epoch in range(EPOCHS):

    print()
    print("=" * 70)
    print(f"Epoch {epoch + 1}/{EPOCHS}")
    print("=" * 70)

    model.train()

    running_loss = 0.0

    optimizer.zero_grad(set_to_none=True)

    progress = tqdm(
        enumerate(train_loader),
        total=len(train_loader),
        leave=True
    )

    start_time = time.time()

    for step, batch in progress:

        batch = {
            k: v.to(DEVICE, non_blocking=True)
            for k, v in batch.items()
        }

        with torch.amp.autocast("cuda"):

            outputs = model(**batch)

            loss = outputs.loss

            loss = loss / GRAD_ACCUM

        scaler.scale(loss).backward()

        running_loss += loss.item() * GRAD_ACCUM

        if (
            (step + 1) % GRAD_ACCUM == 0
            or
            (step + 1) == len(train_loader)
        ):

            scaler.unscale_(optimizer)

            torch.nn.utils.clip_grad_norm_(
                model.parameters(),
                1.0
            )

            scaler.step(optimizer)

            scaler.update()

            scheduler.step()

            optimizer.zero_grad(set_to_none=True)

        progress.set_postfix(
            loss=f"{running_loss/(step+1):.4f}",
            lr=f"{scheduler.get_last_lr()[0]:.2e}"
        )

    epoch_time = time.time() - start_time

    train_loss = running_loss / len(train_loader)

    valid_loss, valid_acc, valid_map3 = validate()

    history["train_loss"].append(train_loss)
    history["valid_loss"].append(valid_loss)
    history["valid_acc"].append(valid_acc)
    history["valid_map3"].append(valid_map3)

    print()

    print(f"Epoch Time      : {epoch_time:.1f} sec")
    print(f"Train Loss      : {train_loss:.4f}")
    print(f"Validation Loss : {valid_loss:.4f}")
    print(f"Validation Acc  : {valid_acc:.4f}")
    print(f"Validation mAP3 : {valid_map3:.4f}")

    if valid_map3 > best_map3:

        best_map3 = valid_map3

        torch.save(
            model.state_dict(),
            BEST_MODEL_PATH
        )

        print("Best model saved.")

print()
print("=" * 70)
print("Training Finished")
print("=" * 70)

print(f"Best Validation mAP@3 : {best_map3:.5f}")
```
